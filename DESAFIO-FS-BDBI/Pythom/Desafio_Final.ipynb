##Questão 2

a. Plote um gráfico de barras (bar) que exiba a quantidade de
funcionários em cada setor (coluna Department);

```python
df2 = df.groupby('Department')['EmployeeNumber'].count()
df2
```

```python
df2.plot(kind = 'bar', color = 'r', title = 'Quantidade de Funcionários por Setores')
plt.xticks(rotation=0)
```

b. Plote um gráfico de linha (line) que exiba a quantidade de
funcionários para cada nível de escolaridade (coluna Education);

```python
df3 = df.groupby('Education')['EmployeeNumber'].count()
df3
```

```python
df3.plot(kind = 'line', color = 'pink', marker='o', title = 'Quantidade de Funcionários por Nível de Escolaridade')
```

c. Plote um gráfico de torta/pizza (pie) que exiba a quantidade de
funcionários que tem ou não algum atrito na empresa (coluna
Attrition).

```python
df4 = df.groupby('Attrition')['EmployeeNumber'].count()
df4
```

```python
df4.plot(kind = 'pie', autopct='%1.1f%%', title = 'Quantidade de Funcionários que possuem Atrito na Empresa')
```
